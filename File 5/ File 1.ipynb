```python
url = 'https://www.imdb.com/list/ls056092300/'
```

```python
def IMDBs_top100_indianmovies(url):
    response = requests.get(url)
    webpage = response.content
    soup = BeautifulSoup(webpage, 'html.parser')

    name = []
    rating = []
    year = []

    movie_containers = soup.find_all('div' , class_='lister-item-content')
    for container in movie_containers:
        names = container.h3.a.text
        name.append(names)

        ratings = container.find('span', class_='ipl-rating-star__rating').text
        rating.append(ratings)

        years = container.find('span', class_='lister-item-year text-muted unbold')
        year.append(years)

    movies_df = pd.DataFrame({
        'Movie': name,
        'Rating': rating,
        'Year': year
    })

    return movies_df
    
movies_top100_df = IMDBs_top100_indianmovies(url)
pd.set_option('display.max_rows',None)
print(movies_top100_df)
    
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from time import sleep
def scrape_bags(url):
    chrome_driver_path = r"C:\Users\shant\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"
    service = Service(executable_path=chrome_driver_path)
    driver = webdriver.Chrome(service=service)
    driver.get(url)
    sleep(5)
    products = driver.find_elements(By.CSS_SELECTOR, ".product-inner")

    data = []
    for product in products:
        title_element = product.find_element(By.CSS_SELECTOR, ".product-title a")
        name = title_element.text

        discounted_price_element = product.find_element(By.CSS_SELECTOR, ".discounted_price")
        discounted_price = discounted_price_element.text if discounted_price_element else "N/A"

        original_price_element = product.find_element(By.CSS_SELECTOR, ".price")
        original_price = original_price_element.text if original_price_element else "N/A"

        data.append({
            "Name": name,
            "Discounted Price": discounted_price,
            "Original Price": original_price
        })

    driver.quit()
    return pd.DataFrame(data)

df1 = scrape_bags(url)
print(df1)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
chrome_driver_path = r"C:\Users\shant\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"
service = Service(executable_path=chrome_driver_path)
driver = webdriver.Chrome(service=service)
def scrape_patreon_post_details(url):
    driver.get(url)
    sleep(2)  
    posts = driver.find_elements(By.CSS_SELECTOR, "div[elevation='subtle']")
    post_details = []

    for post in posts:
        try:
            heading = post.find_element(By.CSS_SELECTOR, "[data-tag='post-title']").text
        except:
            heading = "Heading not available"

        try:
            date = post.find_element(By.CSS_SELECTOR, "[data-tag='post-published-at']").text
        except:
            date = "Date not available"
        
        try:
            content_div = post.find_element(By.CSS_SELECTOR, "div.sc-cfnzm4-0.kJujbw")
            content = content_div.find_element(By.TAG_NAME, "p").text
        except Exception as e:
            print(f"Error extracting content: {e}")
            content = "Content not available or locked"

        try:
            likes = post.find_element(By.CSS_SELECTOR, "[data-tag='like-count']").text
        except:
            likes = "Likes not available"
        
        post_details.append({
            'heading': heading,
            'date': date,
            'content': content,
            'likes': likes
        })

    return post_details

url = "https://www.patreon.com/coreyms/posts"
details = scrape_patreon_post_details(url)

for detail in details:
    print(detail)

driver.quit()
```
