```python
url1 = 'https://www.imdb.com/list/ls091520106'
```

```python
def imdb_top100_movies(url):
    response = requests.get(url)
    webpage = response.content
    soup = BeautifulSoup(webpage, 'html.parser')

    names = []
    ratings = []
    years = []

    movie_containers = soup.find_all('div', class_='lister-item-content')
    for container in movie_containers:
        name = container.h3.a.text
        names.append(name)

        rating = container.find('span', class_='ipl-rating-star__rating').text
        ratings.append(rating)

        year = container.h3.find('span', class_='lister-item-year').text
        years.append(year)

    movies_df = pd.DataFrame({
        'Movie': names,
        'Rating': ratings,
        'Year': years
    })

    return movies_df

movies_top100_df = imdb_top100_movies(url1)
print(movies_top100_df)
```

```python
presidents_urls = {
    "Shri Ram Nath Kovind":"https://ramnathkovind.nic.in/",
   "Shri Pranab Mukherjee": "https://presidentofindia.gov.in/pranab_mukherjee/",
    "Smt Pratibha Devisingh Patil":"https://presidentofindia.gov.in/pratibhadevisinghpatil/",
    "DR. A.P.J. Abdul Kalam":"https://presidentofindia.gov.in/abdulkalam/",
    "Shri K. R. Narayanan":"https://presidentofindia.gov.in/krnarayana/",
     "Dr Shankar Dayal Sharma":"https://presidentofindia.gov.in/Shankar_Dayal/",
    "Shri R Venkataraman":"https://presidentofindia.gov.in/R_Venkataraman/",
     "Giani Zail Singh":"https://presidentofindia.gov.in/Giani_Zail/",
    "Shri Neelam Sanjiva Reddy":"https://presidentofindia.gov.in/Neelam_Sanjiva/",
    "Dr. Fakhruddin Ali Ahmed":"https://presidentofindia.gov.in/Fakhruddin_Ali/",
    "Shri Varahagiri Venkata Giri":"https://presidentofindia.gov.in/Varahagiri_Venkata/",
    "Dr. Zakir Husain":"https://presidentofindia.gov.in/Zakir_Husain/",
     "Dr. Sarvepalli Radhakrishnan": "https://presidentofindia.gov.in/Sarvepalli_Radhakrishnan/",
    "Dr. Rajendra Prasad":"https://presidentofindia.gov.in/Rajendra_Prasad/"
}
```

```python
listing_url = "https://presidentofindia.nic.in/former-presidents"
```

```python
response = requests.get(listing_url)
soup = BeautifulSoup(response.text, 'html.parser')
presidents = soup.select('div.desc-sec > h3')
def get_presidents_names(listing_url):
    response = requests.get(listing_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    presidents = soup.select('div.desc-sec > h3')
    return [president.get_text(strip=True) for president in presidents]

def get_term_of_office(name, presidents_urls):
    url = presidents_urls[name]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    if name == "Shri Ram Nath Kovind":
        term_div = soup.find('div', class_='mini-posts')
        if term_div:
            term_b = term_div.find('b')
            if term_b:
                return term_b.get_text(strip=True)
    else:
        term_container = soup.find('div', class_='brand-text')
        if term_container:
            terms = term_container.find_all('span')
            if len(terms) > 1:
                return terms[1].get_text(strip=True)

    return 'Term not found'

presidents_names = get_presidents_names(listing_url)

presidents_data = []
for name in presidents_names:
    if name in presidents_urls:
        term_of_office = get_term_of_office(name, presidents_urls)
        presidents_data.append({'Name': name, 'Term of Office': term_of_office})
    else:
        print(f"URL for {name} not found in the provided URLs list.")

df = pd.DataFrame(presidents_data)

print(df)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.dineout.co.in/pune-restaurants/welcome-back"

response = requests.get(url)

if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    restaurants = soup.find_all('div', class_='restnt-card restaurant')

    for restaurant in restaurants:
        name_tag = restaurant.find('a', class_='restnt-name')
        location_tag = restaurant.find('div', class_='restnt-loc')
        cuisine_tags = restaurant.find('div', class_='detail-info').find_all('a')
        rating_tag = restaurant.find('div', class_='restnt-rating')
        image_tag = restaurant.find('div', class_='img-wrap').find('src')

        name = name_tag.text if name_tag else "N/A"
        location = location_tag.text if location_tag else "N/A"
        cuisine = ', '.join([cuisine.text for cuisine in cuisine_tags]) if cuisine_tags else "N/A"
        rating = rating_tag.text if rating_tag else "N/A"
        image_url = image_tag.get('src') if image_tag else "N/A"
        if image_url == "N/A": 
            image_url = image_tag.get('data-src') if image_tag else "N/A"
        
        print(f"Name: {name}")
        print(f"Location: {location}")
        print(f"Cuisine: {cuisine}")
        print(f"Rating: {rating}")
        print(f"Image URL: {image_url}")
        print("\n---\n")
else:
    print(f"Failed to retrieve content, status code: {response.status_code}")
```
